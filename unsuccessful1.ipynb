```python
numbers = list(range(1,501))
strnumbers=[str(x) for x in numbers]
url_list=[]
firsturl='https://www.upwork.com/search/profiles/?page=1&q=data%20analyst'
new_url=''
def create_url_list():
    for number in strnumbers:
        listurl=list(firsturl)
        listurl[45]= number
        new_url=''.join(listurl)
        url_list.append(new_url)
    return url_list
```

```python
newlist=(create_url_list())
```

```python
def extract_names():
    big_names=[]
    for link in newlist:
        page = requests.get(link, headers= headers)
        soup1 = soup(page.content, "html.parser")
        soup2 = soup(soup1.prettify(), "html.parser")
        name_list = []
        for name in soup2.findAll('div', {'class':'identity-name'}):
            name_list.append(str(name))
        big_names.append(name_list)
        
    return big_names
```
